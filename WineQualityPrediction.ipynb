```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("winequality_red.csv")
```

```python
df
```

```python
df.describe()

#stats in this 

#25%--25 th percintale
#50 means 50 th persentile data 
```

```python
x=df.drop(columns='quality')
```

```python
y=df['quality']
```

```python
y.unique()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
#using here general holdout method
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=355,test_size=0.30)
```

```python
cls=DecisionTreeClassifier()
#by default taking gini 

cls.fit(x_train,y_train)
cls.score(x_test,y_test)
```

## will see with entropy now

```python
clss=DecisionTreeClassifier(criterion='entropy',max_depth=30,min_samples_leaf=1)
clss.fit(x_train,y_train)
clss.score(x_test,y_test)


#less accuracy
```

## now randomForest

```python
from sklearn.ensemble import RandomForestClassifier
ran=RandomForestClassifier(random_state=6)
#here random state is not about same rows every time ---will select same features and same trees every time we run 
#so will get same accuracy 
ran.fit(x_train,y_train)


#see here by default --10 estimaters---10 tress taking 
```

```python
ran.score(x_test,y_test)

#see here better accuracy with random forest comparing to decision tree
#so ensemble techniques is best 
```

```python
ran=RandomForestClassifier(random_state=6,n_estimators=80)
ran.fit(x_train,y_train)

```

```python
ran.score(x_test,y_test)
#if we took 80 tres better accurac/
```

```python
ran.predict([[7.4,0.700,0.00,1.9,0.076,11.0,34.0,0.99780,3.51,0.56,9.4]])

#we predict correctly classified our model
```

```python
ran=RandomForestClassifier(random_state=6,n_estimators=800)
ran.fit(x_train,y_train)
ran.score(x_test,y_test)

#see here 80 nad 800 trees mostly same bcz --all trees are almost doing the same thing here 
```

## will increase accuracy by doing gridsearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param={
    "n_estimaters":[90,100,115,130],
    'criterion':['gini','entropy'],
    'max_depth':range(2,20,1),
    'min_samples_leaf':range(1,10,1),
    'min_samples_split':range(2,10,1),
    'max_features':['auto','log2']
}
```

```python
grid_search=GridSearchCV(estimator=ran,param_grid=grid_param,cv=5,n_jobs=-1,verbose=3)

#cv---cross validation
#estimater=ran----our model then ding hyper parameter 
```

```python
grid_search.fit(x_train,y_train)
```

```python
ran=RandomForestClassifier(random_state=6,n_estimators=800,max_depth=10,max_features=6)
ran.fit(x_train,y_train)
ran.score(x_test,y_test)
```
